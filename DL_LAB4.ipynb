```python
# import libraries and dataset
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("diabetes.csv")
data.head()
```

```python
# scaling the features using standard scaler 
from sklearn.preprocessing import StandardScaler
scaller=StandardScaler()
data.iloc[:,[0,1]]=scaller.fit_transform(data.iloc[:,[0,1]])
data
```

```python
# splitting into train and test data
from sklearn.model_selection import train_test_split
X,Y=data.iloc[:,[0,1]],data.iloc[:,2]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,)
```

```python
# building a three layer neural network
# apt loss functions are used.
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(16,activation='relu',input_shape=(2,)))
model.add(tf.keras.layers.Dense(16,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
# describe the layers of the model
model.summary()
```

```python
# using adam as the optimizer
opt = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile(loss='binary_crossentropy',optimizer=opt,metrics=['accuracy'])
```

```python
callback=tf.keras.callbacks.EarlyStopping(
    monitor="accuracy",
    min_delta=0,
    patience=100,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
)
```

```python
# compile and fit the model to the training dataset
model.fit(X_train,Y_train,epochs=1000,callbacks=[callback])
```

```python
# get predicitons from the model on the test dataset
ypred=model.predict(X_test)
```

```python
Y_test=np.array(Y_test).reshape(-1,1)
```

```python
# find the mean square error of the model
MSE=sum(((Y_test-ypred)**2))/(2*len(ypred))
```

```python
MSE
```

```python
# using other optimizers
opt = tf.keras.optimizers.SGD(learning_rate=0.002)
model.compile(loss='binary_crossentropy',optimizer=opt,metrics=['accuracy'])

callback=tf.keras.callbacks.EarlyStopping(
    monitor="accuracy",
    min_delta=0,
    patience=100,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
)
```

```python
model.fit(X_train,Y_train,epochs=1000,callbacks=[callback])
```

```python
ypred=model.predict(X_test)
```

```python
Y_test=np.array(Y_test).reshape(-1,1)
```
