```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn import datasets
```

```python
df=pd.read_csv("/content/Iris.csv")
df.head()
```

```python
label=np.array(df.iloc[:,-1])
label=label.reshape(-1,1)
onehot=OneHotEncoder(sparse=False)
y_onehot=onehot.fit_transform(label)
label
y_onehot
```

```python
df1=pd.get_dummies(df.iloc[:,-1:])
df1
```

```python
dfx=pd.concat([df,df1],axis=1)
dfx.drop(["Species","Id"],axis=1,inplace=True)
dfx.head()
```

```python
std=StandardScaler()
dfx.iloc[:,[0,1,2,3]]=std.fit_transform(dfx.iloc[:,[0,1,2,3]])
dfx.head()
```

```python
def sigmoid(x):
  return 1/(1+ np.exp(-x))
```

```python
def msef(a,b):
  return(((a-b)**2).sum())/a.size

def accuracy(a,b):
  acc=a.argmax(axis=1)==b.argmax(axis=1)
  return acc.mean()
```

```python
input=dfx.iloc[:,0:4]
input
```

```python
n=input.shape[0]
n
```

```python
iter=5000
lr=0.01
```

```python
nof=4
hiddenn=2
outputsize=3
```

```python
results1=pd.DataFrame(columns=["mse","accuracy"])
```

```python
np.random.seed(10)
w1=np.random.normal(scale=0.5,size=(nof,hiddenn))
w2=np.random.normal(scale=0.5,size=(hiddenn,outputsize))
bias=np.random.rand(1)
```

```python
for i in range(iter):
  z1=np.dot(input,w1)+bias
  a1=sigmoid(z1)
  z2=np.dot(a1,w2)
  a2=sigmoid(z2)
  mse=msef(a2,np.array(y_onehot))
  acc=accuracy(a2,y_onehot)
  results1=results1.append({"mse":mse,"accuracy":acc},ignore_index=True)
  e1=a2-y_onehot
  dw1=e1*a2*(1-a2)
  e2=np.dot(dw1,w2.T)
  dw2=e2*a1*(1-a1)
  w2_update=np.dot(a1.T,dw1)/n
  w1_update=np.dot(input.T,dw2)/n
  w2=w2-lr*w2_update
  w1=w1-lr*w1_update

  
```

```python
results1
```

```python
results1.mse.plot(title="Mean Square Error")
```

```python
results1.accuracy.plot(title="Accuracy")
```
